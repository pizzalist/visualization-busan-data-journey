```python
regions_review = pd.read_csv("./data/지역별 통계.csv")
```

```python
regions_review[['지역','티켓판매수','티켓판매액']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# create a sample dataframe
df = regions_review[['지역','티켓판매수','티켓판매액']]

sns.barplot(
    data= df,
    x= '지역',
    y= '티켓판매수'
)

plt.rc("font", family= "AppleGothic")

sns.set(font= "AppleGothic",
        rc={"axes.unicode_minus":False},
        style= "darkgrid")

plt.show()
```
